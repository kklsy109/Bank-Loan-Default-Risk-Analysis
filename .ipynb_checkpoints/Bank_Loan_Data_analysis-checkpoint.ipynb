<font size=3 >I have selected 18 attributes that may impact the customer default rate, including NAME_CONTRACT_TYPE, CODE_GENDER, FLAG_OWN_CAR, FLAG_OWN_REALTY, NAME_HOUSING_TYPE, NAME_FAMILY_STATUS, NAME_EDUCATION_TYPE, NAME_INCOME_TYPE, REGION_RATING_CLIENT, OCCUPATION_TYPE, ORGANIZATION_TYPE, FLAG_DOCUMENT_3, AGE_GROUP, EMPLOYMENT_YEAR, AMT_CREDIT_RANGE, AMT_INCOME_RANGE, CNT_CHILDREN, CNT_FAM_MEMBERS.
```python
application_drop_df.head()
```

a) NAME_CONTRACT_TYPE
---

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,10))

# Draw a countplot of the NAME_CONTRACT_TYPE attribute
application_drop_df["NAME_CONTRACT_TYPE"] = application_drop_df["NAME_CONTRACT_TYPE"]

s1 = sns.countplot( ax = ax1,
                    x = application_drop_df["NAME_CONTRACT_TYPE"],
                    hue=application_drop_df["TARGET"].replace({1:"Defaulter",0:"Repayer"}),
                    palette=["#ffb3ec","#8080ff"]
                   )

ax1.set_title("NAME_CONTRACT_TYPE Countplot")

# Draw a barplot of the default rate
default_rate_df = application_drop_df[['NAME_CONTRACT_TYPE', 'TARGET']].groupby(['NAME_CONTRACT_TYPE'],as_index=False).mean()
default_rate_df["TARGET"] = default_rate_df["TARGET"]*100
default_rate_df.sort_values(by='TARGET', ascending=False, inplace=True)

s2 = sns.barplot( ax = ax2,
                 x = 'NAME_CONTRACT_TYPE', 
                 y = 'TARGET', 
                 order = default_rate_df['NAME_CONTRACT_TYPE'], 
                 data = default_rate_df,
                 palette=["#b3d1ff","#ffd1b3"]
               )
plt.ylabel('Percent of Defaulters [%]', fontsize=10)
plt.title("NAME_CONTRACT_TYPE Defaulter")
```

It can be seen that most applicants choose cash loan, and the repayment rate is relatively low.

```python
# Computes correlations between numeric attributes
numerical_corr = numerical_df.corr()
```

```python
# Plot a heat map of correlations between numeric attributes
fig = plt.figure(figsize=(20,20))
ax = sns.heatmap(numerical_corr.corr(), cmap="OrRd",annot=True,linewidth =1)
```

```python
# Top 10 correlation for the numerical data
numerical_corr = numerical_df.corr()
numerical_corr = numerical_corr.where(np.triu(np.ones(numerical_corr.shape),k=1).astype(np.bool))
numerical_corr_df = numerical_corr.unstack().reset_index()
numerical_corr_df.columns =['Attr1','Attr2','Correlation']
numerical_corr_df.dropna(subset = ["Correlation"], inplace = True)
numerical_corr_df["Correlation"]=numerical_corr_df["Correlation"].abs() 
numerical_corr_df.sort_values(by='Correlation', ascending=False, inplace=True) 
numerical_corr_df.head(20)
```

<font size=3 >According to the analysis of the heat map, the number columns related to the amount have highly correlated attributes, and I will draw a line chart to analyze the relationship between them and the customer default rate</font>

```python
# Regroup application data based on target values 0 and 1
Repayer_df = application_drop_df.loc[application_drop_df['TARGET']==0, numerical_corr_list] # Repayers
Defaulter_df = application_drop_df.loc[application_drop_df['TARGET']==1, numerical_corr_list] # Defaulters
```

```python
# Draw line chart
numerical_value = application_drop_df[[ 'AMT_GOODS_PRICE','AMT_CREDIT','CNT_FAM_MEMBERS', 'CNT_CHILDREN','AMT_ANNUITY','DAYS_EMPLOYED',
                                'DAYS_BIRTH']]

fig = plt.figure(figsize=(16,12))

for i in enumerate(numerical_value):
    plt.subplot(4,2,i[0]+1)
    sns.distplot(Defaulter_df[i[1]], hist=False, color='r',label ="Defaulter")
    sns.distplot(Repayer_df[i[1]], hist=False, color='g', label ="Repayer")
    #plt.title(i[1], fontdict={'fontsize' : 10, 'fontweight' : 5, 'color' : 'Blue'}) 
    
plt.legend()
plt.show() 
```

<font size=3 >It can be seen that the repayers and defaulters distributions overlap in all plots, so I cannot use any of these variables in isolation to make a decision.</font>
